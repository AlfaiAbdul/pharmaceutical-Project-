```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib as plt
#import matplotlib.pyplot as plt
import csv
import os
```

# Reading and merging data sets

```python
#extracting ans reading data from source
mouse_drug_data = pd.read_csv('mouse_drug_data.csv')
clinicaltrial_data = pd.read_csv('clinicaltrial_data.csv')
```

```python
clinicaltrial_data.head()
```

```python
mouse_drug_data.head()
```

```python
#merging both set of datas together
merged_data = mouse_drug_data.merge(clinicaltrial_data, on="Mouse ID")
```

```python
merged_data.head()
```

# Measuring response to Treatement 

```python
#tumor val. using drug and timepoint
tumor_vol = merged_data.groupby(['Drug', 'Timepoint']).aggregate({'Tumor Volume (mm3)': 'mean'})
```

```python
tumor_vol
```

```python
timepoint_groupby = merged_data.groupby(['Drug','Timepoint'])
```

```python
#using .sem to get the standard error for the mean at timepoints for 
#each Drug
tumor_Vol_mm3 = timepoint_groupby['Tumor Volume (mm3)'].sem()

tumor_vol_mm3_pd = pd.DataFrame({
    'tumor(Volmm3)' : tumor_Vol_mm3
})

tumor_vol_mm3_pd.head()
```

```python
unpack_tummor_vol = tumor_vol.unstack()
```

```python
unpack_tummor_vol
```

```python
x_axis = unpack_tummor_vol.index

#listing medications
medication = ['Capomulin','Infubinol','Ketapril','Placebo']
markers = ['o','o','o','o']
colors = ['black', 'red','green','blue']

# finding y_axis
for x in medication:
    y = medication.index(x)
    plt.errorbar(x_axis, tumor_vol_mm3_pd['tumor(Volmm3)'][x], yerr=tumor_vol_mm3_pd[x], marker=markers[y], color=colors[y], linestyle='--', label=x)
    
    
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.xlim(unpack_tummor_vol.index[0], unpack_tummor_vol.index[-1])
plt.legend(frameon=True, edgecolor="black")
plt.show()
```
